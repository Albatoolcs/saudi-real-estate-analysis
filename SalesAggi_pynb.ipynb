تجميع ارباع السنه في جدول واحد

```python
all_agg=pd.concat([agg1,agg2,agg3,agg4],axis=0)
all_agg
```

```python
aggregated_sales_by_city_quarter_auction = all_agg.groupby(['المدينة', 'الربع', 'نوع المزاد']).agg(
    total_auctions=('عدد_المزادات', 'sum'),
    total_assets=('عدد_الأصول', 'sum'),
    total_sales_million=('مجموع_قيمة_البيع_مليون', 'sum')
).reset_index()

# Display the result
display(aggregated_sales_by_city_quarter_auction)
```

```python
def highlight_quarter_and_auction(row):
  if row['نوع المزاد'] == 'هجين':
    return ['background-color:lightyellow'] * len(row) # Highlight Hybrid auctions with skyblue
  elif row['الربع'] == 1:
    return ['background-color:lightpink'] * len(row)
  elif row['الربع'] == 2:
    return ['background-color:skyblue'] * len(row)
  elif row['الربع'] == 3:
    return ['background-color:lightblue'] * len(row)
  elif row['الربع'] == 4:
    return ['background-color:lightgreen'] * len(row)
  return [''] * len(row) # Default to no styling if no condition is met

styled_aggregated_sales = aggregated_sales_by_city_quarter_auction.style.apply(highlight_quarter_and_auction, axis=1)
display(styled_aggregated_sales)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate total sales by auction type across all quarters
total_sales_by_auction_type = all_agg.groupby('نوع المزاد')['مجموع_قيمة_البيع_مليون'].sum().reset_index()

# Sort for better visualization
total_sales_by_auction_type = total_sales_by_auction_type.sort_values(by='مجموع_قيمة_البيع_مليون', ascending=False)

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='نوع المزاد', y='مجموع_قيمة_البيع_مليون', data=total_sales_by_auction_type, hue='نوع المزاد', palette='viridis', legend=False)

plt.title('Total Sales Value by Auction Type Across All Quarters (in Millions)')
plt.xlabel('Auction Type (نوع المزاد)')
plt.ylabel('Total Sales Value (Million SAR)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculate the average sales value per auction type for each quarter
average_sales_by_quarter_auction = all_agg.groupby(['الربع', 'نوع المزاد'])['مجموع_قيمة_البيع_مليون'].mean().reset_index()

# Display the result
display(average_sales_by_quarter_auction)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the grouped bar chart
plt.figure(figsize=(12, 7))
sns.barplot(x='الربع', y='مجموع_قيمة_البيع_مليون', hue='نوع المزاد', data=average_sales_by_quarter_auction, palette='viridis')

plt.title('Average Sales Value per Quarter by Auction Type (in Millions)')
plt.xlabel('Quarter')
plt.ylabel('Average Sales Value (Million SAR)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Auction Type')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
